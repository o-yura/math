##### 1. Задание
##### Даны два вектора в трехмерном пространстве: (20,20,20) и (0,0,-20)
##### Найдите их сумму. (на листочке)
##### Напишите код на Python, реализующий расчет длины вектора, заданного его координатами. (в программе)

```python
z = [20, 20, 0]
lenght = 0
for i in z:
    lenght += i ** 2
length = math.sqrt(lenght)
print('Длина вектора:',length)
```

#### 5. Задание (в программе)
#### Нарисуйте трехмерный график двух параллельных плоскостей.
#### Нарисуйте трехмерный график двух любых поверхностей второго порядка.


```python
from mpl_toolkits import mplot3d
```

```python
a = 2
b = 5
d = 10
fig = plt.figure()
ax = plt.axes(projection='3d')
x = np.linspace(10, 20, 100)
y = np.linspace(10, 20, 100)
x, y = np.meshgrid(x, y)
z1 = a * x + b * y
z2 = a * x + b * y + d
ax.contour3D(x, y, z1, 150, cmap='binary')
ax.contour3D(x, y, z2, 100)
plt.show()
```

```python
x = np.outer(np.linspace(-2, 2, 30), np.ones(30))
y = x.copy().T
z1 = np.cos(x ** 2 + y ** 2)
z2 = np.sin(x ** 2 + y ** 2) + 5

fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot_surface(x, y, z1)
ax.plot_surface(x, y, z2)
plt.show()
```
