```python
import matplotlib.pyplot as plt
import numpy as np
from math import exp 
```

1.Решите систему уравнений:

y = x2 – 1

exp(x) + x(1 – y) = 1

```python
x = np.linspace(-3, 5, 200)
y = [(t**2 - 1) for t in x]
y2 = [(exp(t2) + t2 - 1) / t2  for t2 in x]
```

```python
plt.plot(list(x),y, label='y=x^2-1')
plt.plot(list(x),y2, label='exp(x)+x(1-y)=1')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

#### Ответ: Решением является три пересечения графиков.

2. Решите систему уравнений и неравенств:

y = x2 – 1

exp(x) + x(1 – y) > 1

```python
x = np.linspace(-3, 5, 200)
y = [(t**2 - 1) for t in x]
y2 = [(exp(t2) + t2 - 1) / t2  for t2 in x]
```

```python
plt.plot(list(x),y, label='y=x^2-1')
plt.plot(list(x),y2, '--', label='exp(x)+x(1-y)>1')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.fill_between(x, y2, min(x), color='orange', alpha=0.3)
plt.show()
```

#### Ответ: Решением является множество корней на синей кривой в закрашенной области.
